```python
"""
step 1: function iterates through records and find similar words if key word is present
in record, and create new column with the similar words and values.  Return the new 
column as a list of tuples.  currently set to look for medication names
"""

def find_first_similar(list_of_words):
    col=[]
    context=[]
    for row in list_of_words:
        try:
            for word in key_words:
                
                if re.search(str(word),str(row)):
                    col.append(model.wv.similar_by_word(word, topn=5, restrict_vocab=None))
                    context.append(word)
                    continue
                else:
                    col.append(0)
                    context.append(0)
                    continue
        except KeyError as a:
            pass
        except ValueError as e:
            pass
        continue
    return col        
```

```python
"""
funtion adds list of tuples to a dictionary
"""
def add_to_dict(dictionary,list_of_tuples):
    for row in list_of_tuples:
        for label, value in row:
            dictionary[label]=value
            continue
    return dictionary
```

```python
from sklearn.manifold import TSNE
import numpy as np

"""
visulization
"""
keys = ['diabetes','diabetic']
#keys=list(related_terms_dict.keys())
embedding_clusters = []
word_clusters = []
for word in keys:
    embeddings = []
    words = []
    for similar_word, _ in model.most_similar(word, topn=30):
        words.append(similar_word)
        embeddings.append(model[similar_word])
    embedding_clusters.append(embeddings)
    word_clusters.append(words)
```

```python
embedding_clusters = np.array(embedding_clusters)
n, m, k = embedding_clusters.shape
tsne_model_en_2d = TSNE(perplexity=15, n_components=2, init='pca', n_iter=3500, random_state=32)
embeddings_en_2d = np.array(tsne_model_en_2d.fit_transform(embedding_clusters.reshape(n * m, k))).reshape(n, m, 2)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline


def tsne_plot_similar_words(title, labels, embedding_clusters, word_clusters, a, filename=None):
    plt.figure(figsize=(16, 9))
    colors = cm.rainbow(np.linspace(0, 1, len(labels)))
    for label, embeddings, words, color in zip(labels, embedding_clusters, word_clusters, colors):
        x = embeddings[:, 0]
        y = embeddings[:, 1]
        plt.scatter(x, y, c=color, alpha=a, label=label)
        for i, word in enumerate(words):
            plt.annotate(word, alpha=0.5, xy=(x[i], y[i]), xytext=(5, 2),
                         textcoords='offset points', ha='right', va='bottom', size=8)
    plt.legend(loc=4)
    plt.title(title)
    plt.grid(True)
    if filename:
        plt.savefig(filename, format='png', dpi=150, bbox_inches='tight')
    plt.show()


tsne_plot_similar_words('Similar words from i2b2', keys, embeddings_en_2d, word_clusters, 0.7,
                        'similar_words.png')
```

```python
from mpl_toolkits.mplot3d import Axes3D
words_wp = []
embeddings_wp = []
for word in list(model.wv.vocab):
    embeddings_wp.append(model.wv[word])
    words_wp.append(word)
tsne_wp_3d = TSNE(perplexity=30, n_components=3, init='pca', n_iter=3500, random_state=12)
embeddings_wp_3d = tsne_wp_3d.fit_transform(embeddings_wp)

def tsne_plot_3d(title, label, embeddings, a=1):
    fig = plt.figure()
    ax = Axes3D(fig)
    colors = cm.rainbow(np.linspace(0, 1, 1))
    plt.scatter(embeddings[:, 0], embeddings[:, 1], embeddings[:, 2], c=colors, alpha=a, label=label)
    plt.legend(loc=4)
    plt.title(title)
    plt.show()


tsne_plot_3d('Visualizing Embeddings using t-SNE', 'i2b2', embeddings_wp_3d, a=0.1)
```
